```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline = pd.read_csv('J:/Datasets/AirPassengers.csv',index_col="Month")
airline.head()
```

```python
airline.index
```

```python
airline.dropna(inplace = True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
airline.index
```

```python
airline['6-month-SMA'] = airline['#Passengers'].rolling(window = 6).mean()
```

```python
airline['12-month-SMA'] = airline['#Passengers'].rolling(window = 12).mean()
```

```python
airline.plot(figsize = (10, 8))
```

```python
airline['EWMA-12'] = airline['#Passengers'].ewm(span = 12).mean()
airline[['#Passengers','EWMA-12']].plot(figsize=(10,8))
```
